```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating the DataFrame
data = {
    'student_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'attendance': [85, 60, 90, 75, 95],  # Attendance in %
    'mark': [78, 65, 88, 70, 92]         # Marks out of 100
}

df = pd.DataFrame(data)
df
```

```python
# Plotting the correlation
plt.scatter(df['attendance'], df['mark'])
plt.title('Correlation between Attendance and Marks')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks (out of 100)')
plt.grid(True)
plt.show()
```
